```python
# Handling missing values
df.isnull().sum()
```

```python
### filling missing values with the mean or median of the columns
df["Sales_fillna"] = df["Sales"].fillna(df["Sales"].mean)
df
```

```python
## Renaming Columns
df = df.rename(columns={"Date":"Sales_Date"})
df.head()
```

```python
## Change the datatypes
df["Values_Sales"] = df["Value"].fillna(df["Value"].mean()).astype(int)
df.head()
```

```python
## Data Aggrefating and Grouping

grouped_mean = df.groupby("Product")["Value"].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(["Product", "Region"])["Value"].sum()
print(grouped_sum)
```

```python
grouped_agg = df.groupby(["Product", "Region"])["Value"].agg(["mean", "sum", "count"])
print(grouped_agg)
```

```python
### Mergin and joining the dataframes

df1 = pd.DataFrame({"Key": ["A", "B", "C"], "Value": [1, 2, 3]})
print(df1)
df2 = pd.DataFrame({"Key": ["A", "B", "D"], "Value": [1,2,4]})
print(df2)
```

```python
df1
```

```python
df2
```

```python
##Inner merging
pd.merge(df1,df2, on="Key", how="inner")
```

```python
##Outter merging
pd.merge(df1,df2,on="Key", how="outer")
```

```python
## Right merging
pd.merge(df1,df2,on="Key", how="right")
```

```python
## Left merging
pd.merge(df1,df2,on="Key", how="left")
```
