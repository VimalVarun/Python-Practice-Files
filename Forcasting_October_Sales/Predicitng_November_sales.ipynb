```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from datetime import datetime
```

```python
#Question 2. Which product had the highest forecasted sales for the "CND Chennai" sales team in November? 
```

```python
#Load the ELB-Sales_data_set

df = pd.read_excel(r'C:\Users\vimal\OneDrive\Desktop\Python\Forcasting_October_Sales\ELB-Sales-Data.xlsx', parse_dates=['Date'])
df.head()
```

```python
df.columns
```

```python
# Filter the data for "CND Chennai" sales team
chennai_sales = df[df['Sales Team'] == 'CND Chennai']
```

```python
chennai_sales
```

```python
# Group by 'Item Name' and 'Date' to get monthly sales data for each product
chennai_sales['Date'] = pd.to_datetime(chennai_sales['Date'])
chennai_sales_monthly = chennai_sales.groupby([chennai_sales['Item Name'], pd.Grouper(key='Date', freq='M')])['Primary Sales'].sum().reset_index()
```

```python
#Forecast using ARIMA for each product

def forecast_sales_for_product(product_data):
    print(f"Forecasting for product with data:\n{product_data}")
    
    try:
        model = ARIMA(product_data['Primary Sales'], order=(1, 1, 1))
        model_fit = model.fit()
        
        forecast = model_fit.forecast(steps=1)
        
        print(f"Forecast for product: {forecast}")
        
        return forecast.iloc[0]  
    except Exception as e:
        print(f"Error forecasting product: {e}")
        return None 


forecasted_sales = []


products = chennai_sales_monthly['Item Name'].unique()
for product in products:
    product_data = chennai_sales_monthly[chennai_sales_monthly['Item Name'] == product]
    
    
    print(f"Processing product: {product}")
    
    forecasted_value = forecast_sales_for_product(product_data)
    
    
    if forecasted_value is not None:
        forecasted_sales.append({
            'Product': product,
            'Forecasted Sales (November)': forecasted_value
        })


forecast_df = pd.DataFrame(forecasted_sales)


if not forecast_df.empty:
    
    highest_sales_product = forecast_df.loc[forecast_df['Forecasted Sales (November)'].idxmax()]
    print("Product with the highest forecasted sales in November:")
    print(highest_sales_product)
else:
    print("No forecasted sales data available.")
```
