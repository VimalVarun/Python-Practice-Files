```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
#Question 3.What were the forecasted sales of "Britorva 20" in October under Coimbatore HQ? (5 Points) 
```

```python
#Load the ELB-Sales_data_set

df = pd.read_excel(r'C:\Users\vimal\OneDrive\Desktop\Python\Forcasting_October_Sales\ELB-Sales-Data.xlsx', parse_dates=['Date'])
df.head()
```

```python
coimbatore_data = df[(df['HQ'] == 'HQ-Coimbatore') & (df['Item Name'] == 'BRITORVA 20')]
coimbatore_data
```

```python
coimbatore_data['Date'] = pd.to_datetime(coimbatore_data['Date'])
coimbatore_data.set_index('Date', inplace=True)
```

```python
#Sum sales by month to prepare for time-series analysis
monthly_sales = coimbatore_data.resample('M')['Primary Sales'].sum().reset_index()
```

```python
#Build and fit an ARIMA model for forecasting
def forecast_sales(data, month):
    model = ARIMA(data['Primary Sales'], order=(1, 1, 1))
    model_fit = model.fit()
    
    
    forecast = model_fit.forecast(steps=1)  
    return forecast.iloc[0]


october_forecast = forecast_sales(monthly_sales, 'October')


print(f"The forecasted sales of 'Britorva 20' in October under Coimbatore HQ is: {october_forecast:.2f}")
```
